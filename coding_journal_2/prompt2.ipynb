```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(0, 2*math.pi, 100)

def equation(a, b, x):
    return (a * math.e ** (-b * x))
```

```python
y = equation(10, 0.5, x)

plt.xlabel('x')
plt.ylabel('y (with scatter)')

sigmas = np.linspace(2, 0.2, 100)
scatter = np.random.normal(loc=0, scale=sigmas, size=100)
y_new = y + scatter



plt.errorbar(x, y_new, yerr=sigmas, fmt='*', color='pink')

plt.savefig('prompt2_plot.pdf', bbox_inches='tight', dpi=600)
```
