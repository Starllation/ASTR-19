```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(0, 2*math.pi, 100)

def equation(a, b, x):
    return (a * math.e ** (-b * x))
```

```python
y = equation(10, 0.5, x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.savefig('prompt1_plot.pdf', bbox_inches='tight', dpi=600)
```
