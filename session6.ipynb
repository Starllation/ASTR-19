```python
import math
from astropy.table import Table
import numpy as np
```

The following cell defines a function that returns sin of the input as a numpy array.

```python
def sin(x): 
    x_array = np.asarray(x)
    return np.sin(x_array)
```

The following cell defines a function that returns cos of the input as a numpy array.

```python
def cos(x): 
    x_array = np.asarray(x)
    return np.cos(x_array)
```

```python
x = np.linspace(0, 2*math.pi, 1000)
sin_var = sin(x)
cos_var = cos(x)
sin_table = Table([x, sin_var], names=['x', 'sin(x)'])
cos_table = Table([x, cos_var], names=['x', 'cos(x)'])
print(sin_table)
print(cos_table)
```

```python
print(f'{'x':^24} {'sin(x)':^21} {'cos(x)':^21}')
line = '-' * 22
print(f"{line} {line} {line}")

for i in range(10):
    print(f"{x[i]:22} {sin(x[i]):22} {cos(x[i]):22}")
```
